````python
# 패스 능력치 별 평가
def check_skill_pass(x):
  if x >= 90:
    return "우수"
  if x > 70:
    return "보통"
  return "미흡"
````

````python
# 함수 적용하여 새로운 열 추가하기
df_power["패스평가"] = df_power["패스"].apply(check_skill_pass)
df_power.head()
````

## 피봇테이블 - pivot_table
엑셀, 스프레드 시트의 피봇 테이블과 동일합니다.

````python
# 본래 df를 확인
df.head()
````

````python
# '신장'을 index로 지정한 pivot_df 생성하기
# aggfunc : 집계함수(aggregate function)
# aggfunc=np.sum, aggfunc=sum, aggfunc='sum'도 모두 되긴 한다.
pivot_df = pd.pivot_table(df, index="신장",
                          values=['3점슛', '골밑슛'] ,aggfunc='mean')
pivot_df.round(2)

````

## 얕은 복사(Shallow Copy) vs 깊은 복사(Deep Copy)

````python
# 피봇으로 만든 데이터프레임 조회
pivot_df.head()
````

````python
# 1. 얕은 복사 2. 깊은 복사
pivot_df_shallow = pivot_df
pivot_df_copy = pivot_df.copy()
````

````python
# 얕은 복사는 변경이 일어나면 원본도 변경됨
del pivot_df_shallow["골밑슛"]
````

````python
# 얕은 복사는 변경이 일어나면 원본도 변경됨 - 메모리 주소 공유, 사실상 변수명만 다른 것
pivot_df_shallow.head()
pivot_df.head()
````

````python
# 깊은 복사는 별개의 df기에 원본에 영향을 주지 않음
pivot_df_copy.head()
````

````python
# deep copy의 변경을 시켜보면 pivot_df_copy는 변했지만
del pivot_df_copy["3점슛"]
````

````python
# pivot_df는 변하지 않는다.
pivot_df_copy.head()
pivot_df.head()
````
